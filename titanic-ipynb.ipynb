## Step 0: Import Libraries and Load Dataset

We begin by importing pandas and loading the dataset.

```python
import pandas as pd

df = pd.read_csv(files)

print("Dataset loaded successfully ✅")
```

## Step 1: Data Exploration

In this step, we explore the dataset to understand its structure, data types, and summary statistics.

```python
print("\n--- HEAD ---")
df.head()
print("\n--- INFO ---")
df.info()
print("\n--- DESCRIPTION ---")
df.describe()
```

## Step 2: Data Cleaning

We handle missing values and remove unnecessary data:
- Age → filled with median
- Embarked → filled with mode
- Cabin → dropped
- Duplicates → removed

```python
# Fill missing values
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Drop Cabin column
if "Cabin" in df.columns:
    df = df.drop("Cabin", axis=1)

# Remove duplicates
df = df.drop_duplicates()

print("Data cleaned successfully ✅")
```

## Step 3: Data Analysis

We analyze survival patterns using groupby:
- Survival rate by gender
- Survival rate by class
- Average age per class
- Survival rate by age group

```python
df.groupby("Sex")["Survived"].mean()
```

```python
df.groupby("Pclass")["Survived"].mean()
```

```python
df.groupby("Pclass")["Age"].mean()
```

```python
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[0, 12, 18, 60, 100],
    labels=["Child", "Teen", "Adult", "Senior"]
)

df.groupby("AgeGroup", observed=True)["Survived"].mean()
```

## Step 4: Data Filtering

We filter specific groups from the dataset:
- Female survivors
- Child survivors
- First-class survivors

```python
df[(df["Sex"] == "female") & (df["Survived"] == 1)]
```

```python
df[(df["Age"] < 12) & (df["Survived"] == 1)]
```

```python
df[(df["Pclass"] == 1) & (df["Survived"] == 1)]
```

## Step 5: Key Insights

- Females had a higher survival rate than males
- First-class passengers were more likely to survive
- Children had higher survival rates
- The highest survival group was female passengers in 1st class

### Conclusion
This analysis shows that gender, class, and age significantly influenced survival outcomes.
